#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
# your answer here

employee.pivot_table(index=['Department','Title'], values = ['Salary','Name'], aggfunc = {'Name':pd.Series.nunique, 'Salary':np.mean}).reset_index()
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
# your answer here
pivot_Test = employee.pivot_table(index=['Department','Title'], aggfunc = {'Salary':['median', 'min', 'max', 'std']}).reset_index().round(2)
pivot_Test.columns = ['Department','Title', 'MaxSalary', 'MedianSalary', 'MinSalary', 'StdSalary']

pivot_Test
```

```python
# Used Code from Lilit after trying build my own function 
def mean_noMax(col):
    col = col[col != col.max()].mean()
    return round(col,2)
    
employee.pivot_table(index=['Department'], aggfunc = {'Salary':mean_noMax}).reset_index()    
```
